```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
file_sizes = ["20GB", "30GB"]
chunk_sizes = ["16MB", "32MB"]
time_data = np.array([[199.762, 251.104],  # 16MB
                       [239.157, 267.044]]) # 32MB

x = np.arange(len(file_sizes))  # Posiciones en el eje X
width = 0.3  # Ancho de las barras

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, time_data[0], width, label='16MB')
rects2 = ax.bar(x + width/2, time_data[1], width, label='32MB')

# Etiquetas y título
ax.set_xlabel("File Size (GB)")
ax.set_ylabel("Time")
ax.set_title("Tiempo de procesamiento para distintos tamaños de Chunk")
ax.set_xticks(x)
ax.set_xticklabels(file_sizes)
ax.legend(title="Chunk Size")

# Mostrar la gráfica
plt.show()
```
